```python
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
import seaborn as sns
import matplotlib.pyplot as plt
import joblib
from datetime import datetime, timedelta
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Display settings for plots
plt.style.use('ggplot')
```

```python
https://www.kaggle.com/code/allunia/e-commerce-sales
def generate_fashion_data(n_samples=365):  # Generate data for 1 year
    # Generate dates for 1 year
    end_date = datetime.now().replace(hour=0, minute=0, second=0, microsecond=0)
    start_date = end_date - timedelta(days=364)  # 365 days including today
    date_range = pd.date_range(start=start_date, end=end_date, freq='D')

    # Categories for data generation
    seasons = ['Spring', 'Summer', 'Fall', 'Winter']
    cities = ['Mumbai', 'Delhi', 'Bangalore', 'Hyderabad', 'Chennai']
    colors = ['Black', 'White', 'Blue', 'Red', 'Green', 'Yellow', 'Pink', 'Grey']
    clothing_types = ['T-Shirt', 'Hoodie', 'Pants', 'Kurta', 'Dress', 'Jacket']
    genders = ['Male', 'Female']

    # Generate data
    data = {
        'date': date_range,
        'clothing_type': np.random.choice(clothing_types, n_samples),
        'gender': np.random.choice(genders, n_samples),
        'city': np.random.choice(cities, n_samples),
        'color': np.random.choice(colors, n_samples)
    }

    # Assign seasons based on dates
    def get_season(date):
        month = date.month
        if month in [3, 4, 5]:
            return 'Spring'
        elif month in [6, 7, 8]:
            return 'Summer'
        elif month in [9, 10, 11]:
            return 'Fall'
        else:
            return 'Winter'

    data['season'] = [get_season(date) for date in data['date']]

    # Create DataFrame and add features
    df = pd.DataFrame(data)
    df['month'] = df['date'].dt.month
    df['day_of_week'] = df['date'].dt.dayofweek

    # Add price ranges
    price_ranges = {
        'T-Shirt': (15, 50), 'Hoodie': (30, 100), 'Pants': (25, 80),
        'Kurta': (40, 150), 'Dress': (45, 180), 'Jacket': (50, 200)
    }
    df['price_range'] = [np.random.uniform(price_ranges[ct][0], price_ranges[ct][1]) 
                         for ct in df['clothing_type']]

    # Add review scores
    df['review_score'] = np.random.beta(8, 2, n_samples) * 2 + 3  # Scores: 3-5

    # Calculate sales volume with seasonal and weekly effects
    base_sales = np.random.normal(3000, 200, n_samples)
    season_multiplier = {'Winter': 1.3, 'Summer': 1.2, 'Spring': 1.0, 'Fall': 1.1}
    seasonal_effect = np.array([season_multiplier[s] for s in df['season']])
    weekend_effect = np.where(df['day_of_week'].isin([5, 6]), 1.2, 1.0)
    review_effect = 1 + 0.2 * (df['review_score'] - 4.0)

    df['sales_volume'] = (base_sales * seasonal_effect * weekend_effect * review_effect).astype(int)
    return df
```

```python
def preprocess_data(df):
    df_processed = df.copy()

    # Categorical and numerical columns
    categorical_cols = ['clothing_type', 'gender', 'season', 'city', 'color']
    numerical_cols = ['price_range', 'review_score', 'day_of_week', 'month']

    # Encode categorical features
    le_dict = {}
    for col in categorical_cols:
        le_dict[col] = LabelEncoder()
        df_processed[col] = le_dict[col].fit_transform(df_processed[col])

    # Scale numerical features
    scaler = StandardScaler()
    df_processed[numerical_cols] = scaler.fit_transform(df_processed[numerical_cols])

    feature_cols = categorical_cols + numerical_cols
    return df_processed, le_dict, scaler, categorical_cols, numerical_cols, feature_cols
```

```python
# Generate data
df = generate_fashion_data()

# Preprocess data
df_processed, le_dict, scaler, categorical_cols, numerical_cols, feature_cols = preprocess_data(df)

# Define features and target
X = df_processed[feature_cols]
y = df_processed['sales_volume']

# Train model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X, y)
```

```python
# Define the get_season function first
def get_season(date):
    month = date.month
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
    else:
        return 'Winter'

# Generate future data
def generate_future_data(df, days=365):
    """Generate future data for prediction"""
    last_date = df['date'].max()
    future_dates = pd.date_range(start=last_date + timedelta(days=1), 
                                 periods=days, freq='D')
    
    future_data = pd.DataFrame()
    future_data['date'] = future_dates
    future_data['month'] = future_data['date'].dt.month
    future_data['day_of_week'] = future_data['date'].dt.dayofweek
    
    # Add seasons using the get_season function
    future_data['season'] = [get_season(date) for date in future_data['date']]
    
    # Use most common values from historical data for categorical features
    for col in ['clothing_type', 'gender', 'city', 'color']:
        future_data[col] = df[col].mode()[0]
    
    # Use mean values for numerical features
    future_data['price_range'] = df['price_range'].mean()
    future_data['review_score'] = df['review_score'].mean()
    
    return future_data
```

```python
# Generate future data
future_data = generate_future_data(df)
future_processed = future_data.copy()

# Encode and scale future data
for col in categorical_cols:
    future_processed[col] = le_dict[col].transform(future_processed[col])
future_processed[numerical_cols] = scaler.transform(future_processed[numerical_cols])

# Predictions
future_predictions = rf_model.predict(future_processed[feature_cols])

# Plot
plt.figure(figsize=(15, 6))
plt.plot(df['date'], df['sales_volume'], label='Historical Sales', alpha=0.7)
plt.plot(future_data['date'], future_predictions, 'r--', label='12-Month Forecast')
plt.title('Fashion Sales: Historical Data and 12-Month Forecast')
plt.xlabel('Date')
plt.ylabel('Sales Volume')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Save model and preprocessing objects
joblib.dump(rf_model, 'rf_model.pkl')
joblib.dump(le_dict, 'label_encoders.pkl')
joblib.dump(scaler, 'scaler.pkl')
print("Model and preprocessors saved successfully.")
```
